```python
import pandas as pd
import numpy as np
from bertopic import BERTopic
from umap import UMAP
```

```python
import nltk
import string
import re
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
docs = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Dissertation/LinkedInParagraph.xlsx')
#,sep=";", encoding='cp1252',on_bad_lines='skip')
docs
```

```python
x=docs['Context']
x
```

```python
newsletterList=[]
for i in range(len(x)):
  print(i)
  newsletterList.append(docs['Context'][i])
newsletterList
```

```python
import nltk
from nltk.tokenize import sent_tokenize

# Download the NLTK sentence tokenizer data
nltk.download('punkt')

def get_sentences(paragraph):
    # Tokenize the paragraph into sentences
    sentences = sent_tokenize(paragraph)
    return sentences
```

```python
sentences=get_sentences(str(newsletterList))
sentences
```

```python
len(sentences)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer_model = CountVectorizer(stop_words="english")
```

```python
#from bertopic import BERTopic
topic_model = BERTopic(language="english", calculate_probabilities=True, verbose=True)
topics, probs = topic_model.fit_transform(newsletterList)
```

```python
freq = topic_model.get_topic_info();
freq.to_excel('/content/drive/MyDrive/Colab Notebooks/Dissertation/Results/Result_BertLinkedIn_sentence.xlsx')
topic_model.generate_topic_labels()
```

```python
hier_top=topic_model.hierarchical_topics(newsletterList)
topic_model.visualize_hierarchy(hierarchical_topics=hier_top)
```

```python
hier_top.head()
```

```python
topic_model.visualize_barchart()
```

```python

umap_model = UMAP(n_neighbors=15, n_components=5,
                  min_dist=0.0, metric='cosine', random_state=42)
topic_model_umap = BERTopic(umap_model=umap_model, vectorizer_model=vectorizer_model)
```

```python
topics, probs = topic_model_umap.fit_transform(newsletterList)
```
